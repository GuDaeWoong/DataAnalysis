```python
import matplotlib.pyplot as plt
data.groupby(["continent"])["beer_servings","spirit_servings","wine_servings","total_litres_of_pure_alcohol"].sum().plot()
```

```python
#4
data_sum=data.groupby(["continent"])["beer_servings","spirit_servings","wine_servings","total_litres_of_pure_alcohol"].sum()
f,ax=plt.subplots(1,4,figsize=(16,8))
data_sum["beer_servings"].plot.pie(ax=ax[0],autopct="%1.1f%%")
ax[0].set_title("beer_servings")
ax[0].set_ylabel("")

data_sum["spirit_servings"].plot.pie(ax=ax[1],autopct="%1.1f%%")
ax[1].set_title("spirit_servings")
ax[1].set_ylabel("")

data_sum["wine_servings"].plot.pie(ax=ax[2],autopct="%1.1f%%")
ax[2].set_title("wine_servings")
ax[2].set_ylabel("")

data_sum["total_litres_of_pure_alcohol"].plot.pie(ax=ax[3],autopct="%1.1f%%")
ax[3].set_title("total_litres_of_pure_alcohol")
ax[3].set_ylabel("")
```
