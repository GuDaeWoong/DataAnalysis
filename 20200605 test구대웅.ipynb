```python
import pandas as pd
```

```python
data_1=pd.read_excel("../data/test9.xlsx",sheet_name=None)
data_all=pd.concat(data_1,ignore_index=True)
```

```python
import nltk
from konlpy.tag import Twitter; t=Twitter()
```

```python
data_all.KEYWORD
```

```python
present_text=""
for each_line in data_all.KEYWORD:
    present_text=present_text+each_line+"\n"
```

```python
tokens_ko=t.morphs(present_text)
tokens_ko
```

```python
ko=nltk.Text(tokens_ko,name="증후군")
print(len(ko.tokens))
print(len(set(ko.tokens)))
```

```python
ko=nltk.Text(tokens_ko,"증후군")
ko.vocab().most_common()
```

```python
stop_words = [",","\n","G","이","TV","1","4","",";"]
tokens_ko = [ each_word for each_word in tokens_ko if each_word not in stop_words]
```

```python
ko = nltk.Text( tokens_ko,  name='증후군')
ko.vocab().most_common(50)
```

```python
data_all["REGDATE"]=pd.to_datetime(data_all["REGDATE"],format="%Y-%m-%d")
```

```python
data_all=data_all.set_index("REGDATE")
```

```python
data_all.sort_index(ascending=True,inplace=True)
```

```python
data_all["HIT"].plot(figsize=(12,8))
```

```python
data_all_trunc=data_all["2011-03-07":"2016-11-30"]
```

```python
data_all_trunc
```

```python
df2=pd.DataFrame({"ds":data_all_trunc.index,"y":data_all_trunc["HIT"]})
df2.reset_index(inplace=True)
del df2["REGDATE"]
```

```python
df2
```

```python
from fbprophet import Prophet
m=Prophet(daily_seasonality=True,yearly_seasonality=True)
m.fit(df2)
```

```python
future=m.make_future_dataframe(periods=180)
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
m.plot(forecast);
```
