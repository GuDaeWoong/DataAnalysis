```python
시간범위 생성하여 인덱스로 지정하기
```

```python
ebola=pd.read_csv("../data/country_timeseries.csv",parse_dates=[0])
print(ebola.iloc[:5,:5])
```

```python
head_range=pd.date_range(start="2014-12-31",end="2015-01-05")
print(head_range)
```

```python
pd.to_datetime함수를 쓰면 날짜 시간을 나타내는 문자열을 자동으로 datetime자료형으로 바꾼후 datetimeindex자료형 인덱스를 생성한다
pd.date_range함수를 쓰면 모든 날짜 시간 을 일일이 입력할 필요없이 시작일과 종료일 또는 시작일과 기간을 입력하면 범위내의 인덱스를 생성해준다
```

```python
ebola_5=ebola.head()
ebola_5.index=ebola_5["Date"]
ebola_5.reindex(head_range)
```
