melt메서드 인자
id_vars위치를 그대로 유지할 열의 이름을 지정
value_vars:행으로 위치를 변경할 열의 이름을 지정
var_name:value_vars로 위치를 변경할 열의 이름을 지정
value_name:var_name으로 위치를 변경할 열의 데이터를 저장할 열의 이름을 지정

```python
print(pew.iloc[:,0:6])
```

```python
#religion열을 고정으로 피벗(위치를 그대로 유지할 열의 이름)
pew_long=pd.melt(pew,id_vars="religion")
print(pew_long)
```

```python
pew_long=pd.melt(pew,id_vars="religion",var_name="income",value_name="count")
print(pew_long.head())
```

기상데이터의 여러 열을 하나로 정리하기-melt,pivot_table메서드

```python
weather=pd.read_csv("../data/weather.csv")
print(weather.iloc[:5,:11])
```

```python
weather_melt=pd.melt(weather,id_vars=["id","year","month","element"],
                    var_name="day",value_name="temp")
print(weather_melt.head())
```

```python
weather_tidy=weather_melt.pivot_table(index=["id","year","month","day"],
                                     columns="element",
                                     values="temp")
print(weather_tidy)
```

```python
weather_tidy_flat=weather_tidy.reset_index()
print(weather_tidy_flat.head())
```

```python
import pandas as pd
import seaborn as sns
tips=sns.load_dataset("tips")
```

```python
tips["sex_str"]=tips["sex"].astype(str)
print(tips.dtypes)
```

```python
잘못입력한 문자열 처리하기 -to_numeric메서드
```

```python
tips_sub_miss=tips.head(10)
tips_sub_miss.loc[[1,3,5,7],"total_bill"]="missing"
print(tips_sub_miss)
```

```python
print(tips_sub_miss.dtypes)
```

```python
tips_sub_miss["total_bill"].astype(float)
```

```python
pd.to_numeric(tips_sub_miss["total_bill"])
```

```python
#errors="ignore"오류무시 형식은 바뀌지않음
tips_sub_miss["total_bill"]=pd.to_numeric(tips_sub_miss["total_bill"],errors="ignore")
print(tips_sub_miss.dtypes)
```

```python
#errors="coerce"오류무시 형식을 억지로 바꿈
tips_sub_miss["total_bill"]=pd.to_numeric(tips_sub_miss["total_bill"],errors="coerce")
print(tips_sub_miss.dtypes)
```

```python
#downcast="float" 더 작게 만들어줌
tips_sub_miss["total_bill"]=pd.to_numeric(tips_sub_miss["total_bill"],errors="coerce",downcast="float")
print(tips_sub_miss.dtypes)
```
