```python
pweb=pd.read_csv("../data/08. PinkWink Web Traffic.csv",encoding="utf-8",thousands=",",names=["date","hit"],index_col=0)
pweb=pweb[pweb["hit"].notnull()]
pweb.head()
```

```python
pweb["hit"].plot(figsize=(12,4),grid=True)
```

```python
time=np.arange(0,len(pweb))
traffic=pweb["hit"].values
fx=np.linspace(0,time[-1],1000)
```

```python
def error(f,x,y):
    return np.sqrt(np.mean((f(x)-y)**2))
```

```python
fp1=np.polyfit(time,traffic,1)
f1=np.poly1d(fp1)
f2p=np.polyfit(time,traffic,2)
f2=np.poly1d(f2p)
f3p=np.polyfit(time,traffic,3)
f3=np.poly1d(f3p)
f15p=np.polyfit(time,traffic,15)
f15=np.poly1d(f15p)
print(error(f1,time,traffic))
print(error(f2,time,traffic))
print(error(f3,time,traffic))
print(error(f15,time,traffic))
```

```python
plt.figure(figsize=(10,6))
plt.scatter(time,traffic,s=10)
plt.plot(fx,f1(fx),lw=4,label="f1")
plt.plot(fx,f2(fx),lw=4,label="f2")
plt.plot(fx,f3(fx),lw=4,label="f3")
plt.plot(fx,f15(fx),lw=4,label="f15")
plt.grid(True,linestyle="-",color="0.75")
plt.legend(loc=2)
plt.show()
```

```python
df=pd.DataFrame({"ds":pweb.index,"y":pweb["hit"]})
df.reset_index(inplace=True)
df["ds"]=pd.to_datetime(df["ds"],format="%y. %m. %d.")
del df["date"]
#Prophet클래스 객체를 만들고 fit 메서드를 호출
m=Prophet(yearly_seasonality=True,daily_seasonality=True)
m.fit(df);
```

```python
future=m.make_future_dataframe(periods=60)#예측구간 60일 make_future_dataframe:예측날짜 구간 생성
future.tail()
```

```python
forecast = m.predict(future)
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail(3)
                #예측값 예측최소값  예측최대값
```

```python
m.plot(forecast);
```

```python
#선형회귀 및 계절성 성분별로 분리
m.plot_components(forecast);
```
