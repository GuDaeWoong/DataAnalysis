```python
from matplotlib import font_manager, rc
font_name = font_manager.FontProperties(fname = 'c:/Windows/Fonts/malgun.ttf').get_name()
rc('font', family = font_name)
import pandas as pd
import numpy as np
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data1=pd.read_csv('../data/abalone.txt',sep=",",header=None)
```

```python
data1.columns=["Sex","Length","Diameter","Height","Whole weight","Shucked weight","Viscera weight","Shell weight","Rings"]
data1
```

```python
#2
data2=data1[data1["Sex"]!="I"]
data2
```

```python
data2.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
minmaxdata=scaler.fit_transform(data2[["Length","Diameter","Height","Whole weight","Shucked weight","Viscera weight","Shell weight","Rings"]])
```

```python
minmaxdata
```

```python
minmaxdata_1=pd.DataFrame(minmaxdata,
                          columns=["Length","Diameter","Height","Whole weight","Shucked weight","Viscera weight","Shell weight","Rings"])
minmaxdata_1
```

```python
data2
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(minmaxdata_1,data2[["Sex"]],train_size=0.2,
                                               random_state=0)
forest = RandomForestClassifier(random_state=0)
forest.fit(X_train, y_train)
print("훈련 세트 정확도 : {:.2f}".format(forest.score(X_train,y_train)))
print("테스트 세트 정확도 : {:.2f}".format(forest.score(X_test,y_test)))
```

```python
from sklearn.tree import  DecisionTreeClassifier
tree_clf=DecisionTreeClassifier(max_depth=3,random_state=13)
tree_clf.fit(X_train,y_train)
print("훈련 세트 정확도 : {:.2f}".format(forest.score(X_train,y_train)))
print("테스트 세트 정확도 : {:.2f}".format(forest.score(X_test,y_test)))
```

```python
# 특성 중요도
print("특성 중요도 : \n{}".format(forest.feature_importances_))
```

```python
aa=pd.DataFrame(forest.feature_importances_,
                ["Length","Diameter","Height","Whole weight","Shucked weight","Viscera weight","Shell weight","Rings"])
aa.sort_values(0,ascending=False)
```
